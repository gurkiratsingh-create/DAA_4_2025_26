```python
n=int(input("Enter the value of n: "))
board=[["."]*n for _ in range(n)]
res=[]
cols=set()
diag1=set()
diag2=set()
def backtrack(row):
            if row==n:
                res.append(["".join(r) for r in board])
                return
            for col in range(n):
                if col in cols or (row-col) in diag1 or (row+col) in diag2:
                    continue
                board[row][col]="Q"
                cols.add(col)
                diag1.add(row-col)
                diag2.add(row+col)
                backtrack(row+1)
                board[row][col]="."
                cols.remove(col)
                diag1.remove(row-col)
                diag2.remove(row+col)
backtrack(0)
res
```
